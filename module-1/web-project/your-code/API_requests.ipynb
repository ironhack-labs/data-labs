```python
import json
import requests 
import pandas as pd 
pd.set_option('display.max_columns', 500)
import re 
from pandas.io.json import json_normalize 
from math import ceil 
import calendar

#Connecting to API 
url = " https://www.healthcare.gov/api/blog.json" 
response = requests.get(url) 
results_init = response.json()

#Stores response into a dataframe object 
x = json.dumps(results_init) 
y = json.loads(x)
df = pd.DataFrame(data= y['blog']) 


df.head()


```

```python
##Drop columns with NaN, these columns do not contain any values.
df2=df.drop(columns=['metadescripción',
                 'outdated',
                 'outdatedLink',
                 'page_name',
                 'page_type',
                 'page_topic',
                 'page_lifecycle',
                 'page_audience',
                 'page_category',
                 'author',
                 '','order'], axis=1)

##Select blog post by language : 'es' for Spanish or 'en' for English
def getEng(dataframe,language):
    return dataframe.loc[(dataframe['lang'] ==language)]

getEng(df2,'es').head(20)
```

```python
##Convert Date column to 'yyyy-mm-dd' format
def getDate(dataframe,column):
    date_new=[]
    for i in dataframe[column]:
        date=re.findall('\d+-\d+-\d+',i)
        date_new.append(date)
    return date_new

date_lst=getDate(df2,'date')

#Storing the clean values into a new 'Date'columns

df2['date']=date_lst
new_date=[date for x in df2['date'] for date in x]
df2['Date Posted']=new_date
df2.drop("date", axis=1, inplace=True)

```

```python
#Find the words defined in 'key_wrd'in the column 'content'
key_wrd=['enrollment','enroll']
def keyWfreq(column):
    text=[]
    keys=[]
    for w in df2[column]:
        try:
            string=w.split(' ')
            text.append(string)
            for sen in text:
                for word in sen:
                    if word in key_wrd:
                        keys.append(word)
                        df2[column]=word
        except:
            pass
    return keys
    

key_lst=keyWfreq('content')

    
    
```

```python
#Count the frequency of elements in key_wrd in each row under 'Content'
freq=[[x,key_lst.count(x)] for x in set(key_lst)]
print(freq)
```
