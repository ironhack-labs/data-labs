```python
# modules
import json
import requests as r
import pandas as pd
from pandas.io.json import json_normalize
```

```python
# URL for the API
input_1 ='top-headlines'
url = ('https://newsapi.org/v2/everything?'
       'q=Venezuela&'
       'from=2019-09-15&'
       'sortBy=popularity&'
       'apiKey=0f2ab35130164742aa151e2d74427e47')
```

```python
# ".get()" method used from the "requests"
response = r.get(url)
results = response.json()
results
```
