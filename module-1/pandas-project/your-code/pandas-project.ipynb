```python
#Pandas and wget library Import
import pandas as pd
import numpy as np
import re
pd.set_option('display.max_columns',None)

```

```python
#Opens the data file save at defined path 
path= '/Users/mvillavicencio/Documents/Data Analytics/GIT/GitHub_Repo/data-labs/module-1/pandas-project/your-code/'
f=open(path + "GSAF5.csv", "r")

#Reads file and create a DataFrame
data_shark= pd.read_csv(path + "GSAF5.csv", encoding='mac_roman')

#1 Clean white spaces in Column's name
data_shark.columns = data_shark.columns.str.replace(' ', '')

#2 Drops colums from dataframe
def dropNulls(dataframe):
    null_cols = dataframe.isnull().sum()
    drop_cols = list(null_cols[null_cols > 5500].index)
    data_shark1= dataframe.drop(drop_cols, axis=1)
    return data_shark1

data_shark1 = dropNulls(data_shark)


#3 Removes Duplicated rows 
def removeDup(dataframe):
    return dataframe.drop_duplicates()

data_shark2 = removeDup(data_shark1)

#4 Remove strings from Age colums

def removeStr(dataframe,col):
    dataframe[col]=dataframe[col].str.replace('[^\d]','')
    #new_age= pd.to_numeric(dataframe[col])
    return dataframe

data_shark3 = pd.concat([data_shark2, removeStr(data_shark2,'Age')])

#5 Fill NaN in column Age with zeros 
def fillColumn(dataframe,col,val):
    dataframe[col] = dataframe[col].fillna(val)
    return dataframe
data_shark4 =fillColumn(data_shark3,'Age',0)

#6 Change Datatype in column selected
def changeColDtype(data,col):
    data[col]= pd.to_numeric(data[col]) 
    return data

data_shark5=changeColDtype(data_shark4,'Age')


#7 Get names of indexes for which column Age is less than 150
def cleanAge(dataframe,col,value):
    dataframe = dataframe[(dataframe[col] != 0) & (dataframe[col] < value)]
    return dataframe
data_shark6= cleanAge(data_shark5,'Age',150)


#8 Remove rows where sex value is not 'M' or 'F'
def cleanSex(col,dataframe):
    dataframe = dataframe[(dataframe[col] == 'M') | (dataframe[col] == 'F')]
    return dataframe
data_shark7 = cleanSex('Sex',data_shark6)


#9 Group records by columns and calculating mean
def group(dataframe,col):
    return dataframe.groupby([col],as_index = False).mean()


#10 File export
def exportData(dataframe,filename):
    return dataframe.to_csv(filename)

exportData(data_shark7,'export.cvs')
group(data_shark7,'Country').head(10)
```
