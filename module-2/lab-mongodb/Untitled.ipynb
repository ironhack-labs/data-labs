1. All the companies that it's name match 'Babelgum'. Retrieve only their name field.

```python
name_babelgum= db.companies.find({"name": { '$eq': "Babelgum" } } )
name_babelgum[0]['name']
```

2. All the companies that have more than 5000 employees. Limit the search to 20 companies and sort them by number of employees.

```python
num_emp = db.companies.find({"number_of_employees":{"$gt":5000}}).sort("number_of_employees").limit(20)
df = pd.DataFrame(list(num_emp))
display(df)
```

3. All the companies founded between 2000 and 2005, both years included. Retrieve only the name and founded_yearfileds.

```python
comp00_05 = db.companies.find({"founded_year":{"$lte":2005, "$gte":2000} }, projection={"name":1, "founded_year":1})
comp00_05 = pd.DataFrame(list(comp00_05))
comp00_05
```

4. All the companies that had a Valuation Amount of more than 100.000.000 and have been founded before 2010. Retrieve only the name and ipo fields.

```python
valuation_founded = db.companies.find({'$and' : [{"founded_year":{'$lt':2010}}, {"ipo.valuation_amount":{'$gt':100000000}}]},projection={"name": 1 , "founded_year": 1, "ipo.valuation_amount": 1})
valuation_founded = pd.DataFrame(list(valuation_founded))
valuation_founded
```

5. All the companies that have less than 1000 employees and have been founded before 2005. Order them by the number of employees and limit the search to 10 companies.

```python
number_employees = db.companies.find({'$and' : [{"founded_year":{'$lt':2005}}, {"number_of_employees":{"$lt":1000}}]}, projection={"name": 1 , "founded_year": 1, "number_of_employees": 1}).sort("number_of_employees").limit(10)
number_employees = pd.DataFrame(list(number_employees))
number_employees
```

6. All the companies that don't include the partners field.

```python
partners_field = db.companies.find({"partners":{"$eq":[]}}, projection={"name": 1 , "partner_field": 1})
partners_field = pd.DataFrame(list(partners_field))
partners_field
```

7. All the companies that have a null type of value on the category_code field.

```python
category_code_null = db.companies.find({"category_code":{'$type': 10}}, projection={"name": 1 , "category_code": 1})
category_code_null = pd.DataFrame(list(category_code_null))
category_code_null
```

8. All the companies that have at least 100 employees but less than 1000. Retrieve only the name and number of employees fields.

```python
employees = db.companies.find({"number_of_employees":{"$lte":1000, "$gte":100} }, projection={"name":1, "number_of_employees":1})
employees = pd.DataFrame(list(employees))
employees
```

9. Order all the companies by their IPO price descendently.

```python
ipo_v = db.companies.find({"ipo.valuation_amount":{'$type':'number'}}, projection={"name":1 ,"ipo.valuation_amount": 1}).sort("ipo.valuation_amount", py.DESCENDING)
ipo_v = json_normalize(list(ipo_v))
ipo_v
```

10. Retrieve the 10 companies with more employees, order by the number of employees

```python
employees_top10 = db.companies.find(projection={"name": 1 , "number_of_employees": 1, '_id':0}).sort("number_of_employees", py.DESCENDING).limit(10)
employees_top10 = pd.DataFrame(list(employees_top10))
employees_top10
```

11. All the companies founded on the second semester of the year. Limit your search to 1000 companies.

```python
semester = db.companies.find({"founded_month":{"$gte":6, "$lte":12} }, projection={"name":1, "founded_month":1})
semester = json_normalize(list(semester))
semester
```

13. All the companies founded before 2000 that have and acquisition amount of more than 10.000.000

```python
acquisition_amount_10m = db.companies.find({'$and' : [{"founded_year":{'$lt':2000}}, {"acquisition.price_amount":{'$gt':10000000}}]},projection={"name": 1 , "founded_year": 1, "acquisition.price_amount": 1})
acquisition_amount_10m = json_normalize(list(acquisition_amount_10m))
acquisition_amount_10m
```
