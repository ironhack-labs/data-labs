4. Given a list of all 4-letter arrangements that can be formed out of the letters in the word 'ALGORITHM', what are the odds that a randomly selected arrangement will be 'MATH'?

```python
# Your code here
prob = 1 / perm(9, 4) 
  
# Print the obtained permutations 
def odds(prob):
    return prob/(1 - prob)

odds(prob)
```

5. Out of 7 consonants and 4 vowels, how many arragements of 3 consonants and 2 vowels can be formed?

```python
# Your code here

comb(7,3)*comb(4,2)*factorial(5)
```

6. In a group of 6 boys and 4 girls, four children are to be selected. In how many different ways can they be selected such that at least one boy should be there?

```python
# Your code here
boys = sum([comb(6, 4), comb(6, 3)* comb(4, 1), comb(6, 2)* comb(4, 2), comb(6, 1)* comb(4, 3)])
boys
```

8. From a group of 6 men and 7 women, five persons are to be selected to form a committee so that at least 3 women are on the committee. In how many ways can it be done?

```python
# Your code here

at_least_3_women = sum([comb(7, 5), comb(7, 4)* comb(5, 1), comb(7, 3)* comb(5, 2), comb(7, 2)* comb(5, 3)])
at_least_3_women
```

9. A website requires a password consisting of between 5 and 8 alphanumeric characters (A-Za-z0-9). Letters cannot repeat. At least one character must be a letter and at least one character must be a number. How many possible passwords can be created?
