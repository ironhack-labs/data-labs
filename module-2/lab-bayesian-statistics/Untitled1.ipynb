# Principal Component Analysis

```python
import pandas as pd
import numpy as np

np.random.seed(42)
```

```python
r = 100

df = pd.DataFrame({
    'x1': sorted(np.random.rand(r)),
    'x2': sorted(np.random.rand(r)),
    'x3': sorted(np.random.rand(r)),
    'y': np.array(range(r)) + (.5 - np.random.rand(r))
})

df.head()
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['x1'], df['y'], alpha = .5)
plt.scatter(df['x2'], df['y'], color = 'red', alpha = .5)
plt.scatter(df['x3'], df['y'], color = 'green', alpha = .5)
```

```python
from statsmodels.multivariate.pca import PCA

df['pc'] = PCA(df[['x1', 'x2', 'x3']], ncomp = 1).factors  #reduce, componentd from 3 to 1
df['pc'].head()
```

```python
plt.scatter(df['pc'], df['y'], color = 'blue', alpha = .5)
```

```python
import seaborn as sns

sns.regplot(x = 'pc', y = 'y', data  = df)
```
