```python
%matplotlib inline

import matplotlib.pyplot as plt

plt.subplots(1, 2) #n_rows, n_cols
```

```python
mpg = df.groupby('Year')['City MPG', 'Highway MPG'].mean().reset_index()

mpg.head()
```

```python
trend = mpg.query(f"Year == {mpg['Year'].min()} or Year == {mpg['Year'].max()}")

trend
```

```python
plt.title("City vs. Highway MPG")

plt.plot(
    mpg['Year'],
    mpg['City MPG']   
)

plt.plot(
    trend['Year'],
    trend['City MPG'],
    ls = '--',
    lw = 3
)
```

```python
fig, axs = plt.subplots(
    1, #number of col
    2, #number of rows
    figsize = (12,4),
)

axs[0].plot(
    mpg['Year'],
    mpg['City MPG'],
    color = 'red',
)

axs[1].plot(
    mpg['Year'],
    mpg['Highway MPG']   
)

axs[0].legend()
axs[1].legend()

axs[0].set_xlabel('Year')
axs[0].set_ylabel('City MPG')

axs[1].set_xlabel('Year')
axs[1].set_ylabel('Highway MPG')
```
