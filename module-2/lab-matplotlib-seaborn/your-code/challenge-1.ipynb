#### Plot both `y=x^2` and `y=exp(x)`in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
# your code here
from math import exp

fig, axs = plt.subplots(
    1, #number of col
    2, #number of rows
    figsize = (12,4),
)
axs[0].plot(
    x,
    x^2,
    color = 'red',
)

y_1 = [exp(el) for i, el in enumerate(x)]

axs[1].plot(
    x,
    y_1   
)

axs[0].set_xlabel('X')
axs[0].set_ylabel('Y')

axs[1].set_xlabel('X')
axs[1].set_ylabel('Z')

axs[1].set_xscale('log')
axs[1].set_yscale('log')
```
