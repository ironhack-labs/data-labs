#### Import all the libraries that are necessary.

```python
# import libraries here
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### Define data.

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x,y) and (x,z) on the axes.

#### There are 2 ways of doing this. Do in both ways.

*Hint: Check out the `nrows`, `ncols`, and `index` arguments of [subplots](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html)*

#### Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
# your code here-1st way (call `subplots` twice using the `index` parameter)

plt.plot(
    x, # use x as the x-values
    y, # use y as the y-values
    label = 'x vs. y', # label the subplot
)

plt.plot(
    x, # use x as the x-values
    z, # use z as the y-values
    label = 'x vs. z',
)


plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
# Playing around with line width and style
with plt.xkcd(): # used this to for unique line-width and style https://matplotlib.org/gallery/showcase/xkcd.html#sphx-glr-gallery-showcase-xkcd-py
    plt.plot(
    x, # use x as the x-values
    y, # use y as the y-values
    label = 'x vs. y', # label the subplot
)

plt.plot(
    x, # use x as the x-values
    z, # use z as the y-values
    label = 'x vs. z',
    ls = '--', #adjust the line style
    lw = 1, # adjust the line width
)

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
# https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html
# plt.subplot(nrows, ncols, index)
```

```python
# your code here-2st way (call `subplots` only once not using the `index` parameter)
fig, [ax1, ax2] = plt.subplots(1, 2)
ax1.plot(x, y) # in this plot, use x versus y
ax1.set_title('x vs. y') 
ax1.set_xlabel('x')     
ax1.set_ylabel('y')   

ax2.plot(x, z) # in the subplot, use x versus z
ax2.set_title('x vs. z') # title it
ax2.set_xlabel('x') # set x-axis label
ax2.set_ylabel('z') # set y-axis label

plt.show()
```

#### Plot both `y=x^2` and `y=exp(x)`in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
# your code here
from math import exp

plt.plot(
    x, # use x as the x-values
    x ** 2, # y=x^2
    label = 'y=x^2', # label the subplot
)

plt.plot(
    x, # use x as the x-values
#     np.exp(x), # y=exp(x) # before class, was using numpy.exp for this one
    [exp(value) for value in x], # from math library, but had to use a listcomp b/c exp() is expecting a single value
#     int(exp(x)), # from math
    label = 'y=exp(x)',
)

# plt.set_xscale = 'linear'
# plt.set_yscale = 'log'

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
