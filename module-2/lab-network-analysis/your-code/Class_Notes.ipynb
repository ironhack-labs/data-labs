# Network Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
```

## Nodes

```python
G = nx.Graph()

G.add_node('A')
G.add_nodes_from(['A', 'B', 'C'])

nx.draw(
    G,
    with_labels = True,
    font_weight = 'bold'
)
```

## Edges

```python
G = nx.Graph()

# G.add_edge('A', 'B')
G.add_edges_from([['A', 'B'], ['B', 'C']])

nx.draw(
    G,
    with_labels = True,
    font_weight = 'bold'
)
```

```python
from itertools import combinations

combos = list(combinations(range(10), 2))
combos[:5]
```

## `nx.draw()`

```python
G = nx.Graph()

G.add_edges_from(combos)

nx.draw(G)
```

```python
G = nx.Graph()

G.add_edges_from(combos)

nx.draw_kamada_kawai(
    G,
    alhpa = 1,
    with_labels = True,
    node_size = 200
)
```

```python
G = nx.Graph()

a = list(combinations(range(0, 5), 2))
b = list(combinations(range(4, 10), 2))

G.add_edges_from(a + b)

nx.draw(
    G,
    alhpa = 1,
    with_labels = True,
    node_size = 200
)
```

## Density

Density is the percentage of the graph that can be reached by the average node.

```python
nx.density(G)
```

## Betweenness

Betweeness centrality tells us which nodes in our network are likely pathways for information.

```python
nx.betweenness_centrality(
    G,
    weight = 'edge'
)
```

## Closeness

Closeness centrality measures node reach or how fast information would spread from that node to other nodes

```python
nx.closeness_centrality(G, distance ='edge')
```

## Eigenvector

Eigencevtor centrality measures related influence or who is closest to the most important nodes in the network.

```python
nx.eigenvector_centrality_numpy(G)
```

## Degree centrality

Degree centrality is a measures of popularity based on a node's degree.

```python
nx.degree_centrality(G)
```

## PageRank centrality

PageRank centrality is a variant of Eigenvector centrality that uses edges from other important nodes as a measure of a node's importance.

```python
nx.pagerank(G)
```
