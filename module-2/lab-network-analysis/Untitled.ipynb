# NODES

```python
G = nx.Graph() #initialize seassion, 
    
# G.add_node('A')
G.add_nodes_from(['A', 'B', 'C'])

nx.draw( #to visualize above code
    G,
    with_labels = True, #to always show the name of the node
    font_weight = 'bold'  
)
```

# EDGES

```python
G = nx.Graph() #initialize seassion, 
    
# G.add_edge('A', 'B')
G.add_edges_from([('A', 'B'), ('B', 'C'), ('C', 'A')])

nx.draw( #to visualize above code
    G,
    with_labels = True, #to always show the name of the node
    font_weight = 'bold'  
)
```

```python
from itertools import combinations

combos = list(combinations(range(10), 2))
combos[:5]
```

```python
G = nx.Graph() #initialize seassion, 
    
a = list(combinations(range(0, 5), 2))
b = list(combinations(range(4, 10), 2))
# G.add_edge('A', 'B')
G.add_edges_from(a+b)

nx.draw( #to visualize above code
    G,
#     alpha = 1, #oppacity
    with_labels = True, #to always show the name of the node
#     node_size = 200  
)
```

# Average Degree

```python
G.degree() #will tell you how many nodes the first number is connected to
```

```python
dict(G.degree()).values()
```

```python
dict(G.degree()).keys()
```

```python
sum(dict(G.degree()).values())/G.order()  #average degree of separation
```

# Betweenness

Betweenness centrality tells us which nodes in our network are likely pathways for information.

```python
nx.betweenness_centrality(
    G, 
#     weight = 'edge'
)
```

# Closeness

Closeness centrality measures node reach or how fast information would spread from that node to the other nodes

```python
nx.closeness_centrality(
    G, 
    distance = 'edge'
)
```

# Eigenvector

Measures related influence or who is closest to the most important nodes in the network

```python
nx.eigenvector_centrality_numpy(
    G, 
#     weight = 'edge'
)
```

# Degree Centrality

Measures popularity based on a node's degree

```python
nx.degree_centrality(
    G, 
#     weight = 'edge'
)
```

# PageRank Centrality

Variant of Eigenvector that uses edges from other important nodes as a measure of a node's importance.

```python
nx.pagerank(G)
```
